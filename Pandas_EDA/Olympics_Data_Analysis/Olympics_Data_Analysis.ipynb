```python
# Total participants per year
participants_per_year = df.groupby('Year')['Name'].count()

# Medal winners per year (drop NaN medals first)
winners_per_year = df.dropna(subset=['Medal']).groupby('Year')['Name'].count()

# Combine into one DataFrame
yearly_stats = pd.DataFrame({
    'Participants': participants_per_year,
    'Winners': winners_per_year
}).fillna(0)   # fill years with 0 winners if none

print(yearly_stats.head(10))  # first 10 years
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.bar(participants_per_year.index, participants_per_year.values, 
        label='Participants', alpha=0.6, color='blue')
plt.bar(winners_per_year.index, winners_per_year.values, 
        label='Winners', alpha=0.6, color='green')

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Participants vs Medal Winners per Year')
plt.legend()
plt.show()
```
